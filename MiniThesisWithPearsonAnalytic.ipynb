```python
import pandas as pd
from scipy.stats import spearmanr
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = "/Data_Dummy.xlsx"
df = pd.read_excel(file_path)

# Ambil dua kolom yang akan diuji
hours = df["Hours_on_Social_Media_per_Day"]
productivity = df["Self_Rated_Productivity"]

print("=== UJI KORELASI SPEARMAN ===")
print("Variabel X: Hours_on_Social_Media_per_Day")
print("Variabel Y: Self_Rated_Productivity")
print()

# Statistik deskriptif
print("STATISTIK DESKRIPTIF:")
print(f"Hours_on_Social_Media_per_Day:")
print(f"  Mean: {hours.mean():.2f}")
print(f"  Std:  {hours.std():.2f}")
print(f"  Min:  {hours.min():.2f}")
print(f"  Max:  {hours.max():.2f}")
print()
print(f"Self_Rated_Productivity:")
print(f"  Mean: {productivity.mean():.2f}")
print(f"  Std:  {productivity.std():.2f}")
print(f"  Min:  {productivity.min():.2f}")
print(f"  Max:  {productivity.max():.2f}")
print()

# Uji Korelasi Spearman
correlation_coeff, p_value = spearmanr(hours, productivity)

print("=== HASIL UJI KORELASI SPEARMAN ===")
print(f"Koefisien Korelasi Spearman (ρ): {correlation_coeff:.4f}")
print(f"P-value: {p_value:.6f}")
print()

# Interpretasi koefisien korelasi
if abs(correlation_coeff) < 0.1:
    strength = "sangat lemah"
elif abs(correlation_coeff) < 0.3:
    strength = "lemah"
elif abs(correlation_coeff) < 0.5:
    strength = "sedang"
elif abs(correlation_coeff) < 0.7:
    strength = "kuat"
else:
    strength = "sangat kuat"

if correlation_coeff > 0:
    direction = "positif"
elif correlation_coeff < 0:
    direction = "negatif"
else:
    direction = "tidak ada"

print("=== INTERPRETASI ===")
print(f"Kekuatan korelasi: {strength}")
print(f"Arah korelasi: {direction}")
print()

# Kesimpulan hipotesis
alpha = 0.05
print("=== KESIMPULAN HIPOTESIS ===")
print(f"H₀: Tidak ada hubungan antara jam penggunaan media sosial dan produktivitas")
print(f"H₁: Ada hubungan antara jam penggunaan media sosial dan produktivitas")
print(f"α = {alpha}")
print()

if p_value < alpha:
    print(f"Keputusan: Tolak H₀ (p-value = {p_value:.6f} < α = {alpha})")
    print(f"Kesimpulan: Ada hubungan yang signifikan secara statistik antara jam penggunaan media sosial dan produktivitas akademik")
    print(f"Hubungan tersebut bersifat {direction} dengan kekuatan {strength}")
else:
    print(f"Keputusan: Gagal tolak H₀ (p-value = {p_value:.6f} > α = {alpha})")
    print(f"Kesimpulan: Tidak ada hubungan yang signifikan secara statistik antara jam penggunaan media sosial dan produktivitas akademik")

# Visualisasi scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(hours, productivity, alpha=0.6, color='blue')
plt.xlabel('Hours on Social Media per Day')
plt.ylabel('Self Rated Productivity')
plt.title(f'Scatter Plot: Media Sosial vs Produktivitas\n(Spearman ρ = {correlation_coeff:.4f}, p = {p_value:.6f})')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

# Tabel korelasi
correlation_matrix = df[['Hours_on_Social_Media_per_Day', 'Self_Rated_Productivity']].corr(method='spearman')
print("\n=== MATRIKS KORELASI SPEARMAN ===")
print(correlation_matrix.round(4))
```

```python
import pandas as pd
import numpy as np
from scipy.stats import pearsonr
import matplotlib.pyplot as plt
import seaborn as sns

# Load data dari file yang sudah dibuat
try:
    # Jika menggunakan file Excel
    df_normal = pd.read_excel("/Data_Dummy_Normal11.xlsx")
except:
    # Jika menggunakan data yang sudah ada di memory
    hours = df_normal["Hours_on_Social_Media_per_Day"]
    productivity = df_normal["Self_Rated_Productivity"]
else:
    hours = df_normal["Hours_on_Social_Media_per_Day"]
    productivity = df_normal["Self_Rated_Productivity"]

print("=== UJI KORELASI PEARSON ===")
print("Variabel X: Hours_on_Social_Media_per_Day")
print("Variabel Y: Self_Rated_Productivity")
print(f"Ukuran sampel (n): {len(hours)}")
print()

# Statistik deskriptif
print("STATISTIK DESKRIPTIF:")
print(f"Hours_on_Social_Media_per_Day:")
print(f"  Mean: {hours.mean():.2f}")
print(f"  Std:  {hours.std():.2f}")
print(f"  Min:  {hours.min():.2f}")
print(f"  Max:  {hours.max():.2f}")
print()
print(f"Self_Rated_Productivity:")
print(f"  Mean: {productivity.mean():.2f}")
print(f"  Std:  {productivity.std():.2f}")
print(f"  Min:  {productivity.min():.2f}")
print(f"  Max:  {productivity.max():.2f}")
print()

# Uji Korelasi Pearson
correlation_coeff, p_value = pearsonr(hours, productivity)

print("=== HASIL UJI KORELASI PEARSON ===")
print(f"Koefisien Korelasi Pearson (r): {correlation_coeff:.4f}")
print(f"P-value: {p_value:.6f}")
print(f"Koefisien Determinasi (r²): {correlation_coeff**2:.4f}")
print(f"Persentase varians yang dijelaskan: {(correlation_coeff**2)*100:.2f}%")
print()

# Interpretasi koefisien korelasi
if abs(correlation_coeff) < 0.1:
    strength = "sangat lemah"
elif abs(correlation_coeff) < 0.3:
    strength = "lemah"
elif abs(correlation_coeff) < 0.5:
    strength = "sedang"
elif abs(correlation_coeff) < 0.7:
    strength = "kuat"
else:
    strength = "sangat kuat"

if correlation_coeff > 0:
    direction = "positif"
elif correlation_coeff < 0:
    direction = "negatif"
else:
    direction = "tidak ada"

print("=== INTERPRETASI ===")
print(f"Kekuatan korelasi: {strength}")
print(f"Arah korelasi: {direction}")
print()

# Kesimpulan hipotesis
alpha = 0.05
print("=== KESIMPULAN HIPOTESIS ===")
print(f"H₀: Tidak ada hubungan linear antara jam penggunaan media sosial dan produktivitas")
print(f"H₁: Ada hubungan linear antara jam penggunaan media sosial dan produktivitas")
print(f"α = {alpha}")
print()

if p_value < alpha:
    print(f"Keputusan: Tolak H₀ (p-value = {p_value:.6f} < α = {alpha})")
    print(f"Kesimpulan: Ada hubungan linear yang signifikan secara statistik antara jam penggunaan media sosial dan produktivitas akademik")
    print(f"Hubungan tersebut bersifat {direction} dengan kekuatan {strength}")
else:
    print(f"Keputusan: Gagal tolak H₀ (p-value = {p_value:.6f} > α = {alpha})")
    print(f"Kesimpulan: Tidak ada hubungan linear yang signifikan secara statistik antara jam penggunaan media sosial dan produktivitas akademik")

# Visualisasi
fig, axes = plt.subplots(2, 2, figsize=(15, 12))

# 1. Scatter plot dengan regression line
axes[0,0].scatter(hours, productivity, alpha=0.6, color='blue')
z = np.polyfit(hours, productivity, 1)
p = np.poly1d(z)
axes[0,0].plot(hours, p(hours), "r--", alpha=0.8)
axes[0,0].set_xlabel('Hours on Social Media per Day')
axes[0,0].set_ylabel('Self Rated Productivity')
axes[0,0].set_title(f'Scatter Plot dengan Garis Regresi\n(Pearson r = {correlation_coeff:.4f}, p = {p_value:.6f})')
axes[0,0].grid(True, alpha=0.3)

# 2. Histogram Hours
axes[0,1].hist(hours, bins=15, alpha=0.7, color='skyblue', edgecolor='black')
axes[0,1].set_xlabel('Hours on Social Media per Day')
axes[0,1].set_ylabel('Frequency')
axes[0,1].set_title('Distribusi Hours on Social Media')
axes[0,1].grid(True, alpha=0.3)

# 3. Histogram Productivity
axes[1,0].hist(productivity, bins=15, alpha=0.7, color='lightgreen', edgecolor='black')
axes[1,0].set_xlabel('Self Rated Productivity')
axes[1,0].set_ylabel('Frequency')
axes[1,0].set_title('Distribusi Self Rated Productivity')
axes[1,0].grid(True, alpha=0.3)

# 4. Heatmap korelasi
correlation_matrix = df_normal[['Hours_on_Social_Media_per_Day', 'Self_Rated_Productivity']].corr()
im = axes[1,1].imshow(correlation_matrix, cmap='coolwarm', aspect='auto', vmin=-1, vmax=1)
axes[1,1].set_xticks([0, 1])
axes[1,1].set_yticks([0, 1])
axes[1,1].set_xticklabels(['Hours', 'Productivity'], rotation=45)
axes[1,1].set_yticklabels(['Hours', 'Productivity'])
axes[1,1].set_title('Heatmap Korelasi')

# Tambahkan nilai korelasi di heatmap
for i in range(2):
    for j in range(2):
        text = axes[1,1].text(j, i, f'{correlation_matrix.iloc[i, j]:.3f}',
                             ha="center", va="center", color="black", fontweight='bold')

plt.colorbar(im, ax=axes[1,1])
plt.tight_layout()
plt.show()

# Tabel korelasi
print("\n=== MATRIKS KORELASI PEARSON ===")
print(correlation_matrix.round(4))

# Interval kepercayaan untuk korelasi (approximation)
import scipy.stats as stats
n = len(hours)
r = correlation_coeff
fisher_z = 0.5 * np.log((1 + r) / (1 - r))
se = 1 / np.sqrt(n - 3)
z_critical = stats.norm.ppf(0.975)  # untuk 95% CI
ci_lower_z = fisher_z - z_critical * se
ci_upper_z = fisher_z + z_critical * se
ci_lower = (np.exp(2 * ci_lower_z) - 1) / (np.exp(2 * ci_lower_z) + 1)
ci_upper = (np.exp(2 * ci_upper_z) - 1) / (np.exp(2 * ci_upper_z) + 1)

print(f"\n=== INTERVAL KEPERCAYAAN 95% ===")
print(f"Korelasi Pearson: {correlation_coeff:.4f}")
print(f"95% CI: [{ci_lower:.4f}, {ci_upper:.4f}]")
```
